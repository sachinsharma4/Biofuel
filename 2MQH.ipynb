Importing

```python
from Bio.Seq import Seq
from Bio.PDB import PDBParser,MMCIFParser
import urllib
```

Set the pdb Id of protein of interest

```python
pdb_id='2MQH'
```

A function for downloading protein structure using pdb Id

```python
def pdb_download(x) :
     string='https://files.rcsb.org/download/' +x+'.pdb'
     urllib.request.urlretrieve(string,x+'.pdb' )
```

Adjustment for Google colabratoy

Importing nglview and testing it

```python
import nglview as nv
nv.demo()
```

Downloading and preparing the structure

```python


pdb_download(pdb_id)
parser = PDBParser()
structure = parser.get_structure(pdb_id, pdb_id+'.pdb')
structure
```

Viewing the structure

```python
view = nv.show_biopython(structure)
view
```

Take image snap shot wait for few while image loads

```python
image_bin=view.render_image()
image_bin
```

Please run this cell only after previous cell has loaded the completely

```python
f = open(pdb_id +".png", "wb")
f.write(image_bin.value)
f.close()
```
